### Data Analysis and visulization 

```python
cancelled_perc = df['is_canceled'].value_counts(normalize=True)
cancelled_perc
```

```python
plt.figure(figsize=(5,5))
plt.title('Reservation Status Count')
plt.bar(['Not cancelled','cancelled'],df['is_canceled'].value_counts(),edgecolor = 'k',width =0.7)
plt.show()
```

```python
resort_hotel = df[df['hotel'] =='Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
city_hotel = df[df['hotel']=='City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
plt.figure(figsize=(8,4))
ax1 = sns.countplot(x='hotel',hue='is_canceled',data=df,palette='gray')
#legend_labels,_ = ax1.get_legend_handles_labels()
plt.title('Hotel-vise Reservation Status', size = 20)
plt.xlabel('Hotel')
plt.ylabel('Number of Reservations')
plt.show()
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(20,8))
plt.title('Average Daily Rate in City and Resort Hotel', fontsize=30)
plt.plot(resort_hotel.index,resort_hotel['adr'],label='Resort Hotel')
plt.plot(city_hotel.index,city_hotel['adr'],label='City Hotel')
plt.legend(fontsize=20);
```

```python
df['month'] = df['reservation_status_date'].dt.month
plt.figure(figsize=(16,8))
ax1 = sns.countplot(x = 'month',data=df,hue='is_canceled')
plt.title('Reservation Status per Month',fontsize=20)
plt.xlabel('Month')
plt.ylabel('Number of Reservations')
plt.legend(['Not Canceled','Canceled']);
 
```

```python
plt.figure(figsize=(15,8))
plt.title('ADR per month',fontsize=30)
sns.barplot(x='month', y='adr', data=df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index())
plt.show()
```

```python
canceled_data = df[df['is_canceled']==1]
top_10_country = canceled_data['country'].value_counts()[:10]
plt.figure(figsize=(6,6))
plt.title('Top 10 Country with Reservation Cancelled')
plt.pie(top_10_country,autopct='%.2f', labels=top_10_country.index);
```

```python
df['market_segment'].value_counts()
```

```python
df['market_segment'].value_counts(normalize=True)
```

```python
canceled_data['market_segment'].value_counts(normalize=True)
```

```python
canceled_df_adr = canceled_data.groupby('reservation_status_date')[['adr']].mean()
canceled_df_adr.reset_index(inplace=True)
canceled_df_adr.sort_values('reservation_status_date',inplace=True)

non_canceled_data = df[df['is_canceled']==0]
non_canceled_df_adr = non_canceled_data.groupby('reservation_status_date')[['adr']].mean()
non_canceled_df_adr.reset_index(inplace=True)
non_canceled_df_adr.sort_values('reservation_status_date',inplace=True)
```

```python
plt.figure(figsize=(20,6))
plt.title('Average Daily Rate',fontsize=30)
plt.plot(non_canceled_df_adr['reservation_status_date'],non_canceled_df_adr['adr'],label='non cancelled')
plt.plot(canceled_df_adr['reservation_status_date'],canceled_df_adr['adr'],label='cancelled')
plt.legend(fontsize=20);
```
